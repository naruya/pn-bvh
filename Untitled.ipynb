```python
import numpy as np


class PNBVH(object):
    def __init__(self, file):
        with open(file) as f:
            bvh = f.readlines()

        _l = bvh.index('MOTION\n')
        self.hierarchy = Hierarchy(bvh[1:_l])
        self.motion = Motion(bvh[_l+4:])

        self.T = bvh[_l+2].split()[-1]
        self.F = 1.0 / float(bvh[_l+3].split()[-1])
        print(self.T, self.F)


class Motion(object):
    def __init__(self, data):
        self.data = np.array([a.split() for a in data], dtype=np.float)


class Hierarchy(object):
    def __init__(self, data):
        self.nodes = []
        self.parse(data)
        self.register_nodes()

    def parse(self, data):
        data = [a.split() for a in data]
        idx = 0
        p_idx = -1  # parent
        for i, a in enumerate(data):
            if a[0] == '{':
                node = Node(data[i-1:i+3], idx, p_idx)
                self.nodes.append(node)
                p_idx = idx
                idx += 1
            if a[0] == '}':
                p_idx = self.nodes[p_idx].p_idx

    def register_nodes(self):
        # register children
        for node in self.nodes[1:]:
            parent = self.nodes[node.p_idx]
            parent.c_idxs.append(node.idx)

        # register parent/children pointer
        for node in self.nodes:
            parent = self.nodes[node.p_idx]
            children = [self.nodes[c_idx] for c_idx in node.c_idxs]
            node._register(parent, children)

    def _print_children(self, node, space):
        print(space + "- " + node.name)
        for child in node.children:
            self._print_children(child, space + " ")

    def print_hierarchy(self):
        self._print_children(self.nodes[0], "")


class Node(object):
    def __init__(self, data, idx, p_idx):
        self.idx = idx
        self.p_idx = p_idx
        self.c_idxs = []
        self.type = data[0][0]
        self.name = data[0][1]
        self.offset = np.array(data[2][1:], dtype=np.float)
        self.channels = data[3][2:]
        self.is_root = (self.type == "ROOT")
        self.is_end = (self.type == "End")
        self.p = None

    def _register(self, parent, children):
        self.parent = parent
        self.children = children
```

```python
bvh = PNBVH('Scene_10Char00.bvh')
```

```python
print(len(bvh.hierarchy.nodes))
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

x, y, z = [], [], []

for node in bvh.hierarchy.nodes:
    if node.is_root:
        node.p = node.offset
    else:
        node.p = node.parent.p + node.offset
    x.append(node.p[0])
    y.append(node.p[1])
    z.append(node.p[2])

fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlim(-100,100)
ax.set_ylim(-100,100)
ax.set_zlim(0,200)
ax.scatter(x, z, y)
ax.view_init(elev=None, azim=120)
plt.show()
```
